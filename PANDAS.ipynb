```python
filter_mask = (df['Hobbyist'] == 'Yes') #Parenthesis not required
filter_mask
```

```python
df[filter_mask] #Returns only those dataframe objects with filter_mask True
```

```python
#Alternatively, use df.loc to filter using the filter_mask. Also, you can add other columns to return as you'd like.
df.loc[filter_mask, 'OpenSourcer']
```

```python
# & --> AND
# | --> OR

# You can combine filter masks
filt = (df['Hobbyist'] == 'Yes') & (df['Student'] == 'No')
df.loc[filt, ['Employment', 'LanguageWorkedWith']]
```

```python
# To return dataframe objects with filter mask False, use a tilde ~
df.loc[~filt, ['Employment', 'Country']]
```

```python
countries = ['India', 'Nepal']
country_filter = df['Country'].isin(countries)
df.loc[country_filter, ['Country', 'ConvertedComp']]
```

```python
language_filter = df['LanguageWorkedWith'].str.contains('Python', na=False) #na makes sure the condition disregards NaN values
df.loc[language_filter, ['LanguageWorkedWith', 'Employment']]
```

```python
# Renaming columns
df.columns = [x.upper() for x in df.columns]
df
```

```python
df.rename(columns={'HOBBYIST': 'HOBBY', 'EDLEVEL': 'EDUCATION'}, inplace=True)
df
```

```python
df.loc[0, ['EDUCATION', 'UNDERGRADMAJOR']] = ['Bachelor’s degree (BA, BS, B.Eng., etc.)', 'Web development or web design']
df.head()
```

```python
df['HOBBY'] = df['HOBBY'].str.lower() # Lowercases all responses in HOBBY
df.head()
```

```python
# apply function
# Enables the use of functions and methods in a data frame

def uppercase_all(column_name):
    return column_name.upper()

df['HOBBY'] = df['HOBBY'].apply(uppercase_all)
df.head()
```

```python
# applymap method
# Applies a function to each element in a dataframe

# df.applymap(len)
# df.applymap(str.lower)
```

```python
# map method
# Substitutes each value in a series with another value

df['HOBBY'] = df['HOBBY'].map({'YES': 'YES I DO PROGRAM AS A HOBBY', 'NO': 'NO I DO NOT PROGRAM AS A HOBBY'})
```

```python
# Making a new column 

df['SUBJECT OF EXPERTISE'] = df['EDUCATION'] + ' in ' + df['UNDERGRADMAJOR']
df.head()
```

```python
# Remove columns

df.drop(columns=['SUBJECT OF EXPERTISE'], inplace=True)
```
